```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Time_series_analysis_and_forecast_DATASET.csv')
```

```python

df.head(20)
```

```python
df["DateTimeCol"]=pd.to_datetime(df['FullDate'])
```

```python
df["date"] = df["DateTimeCol"].dt.date
df["time"] = df["DateTimeCol"].dt.time
```

```python
df.drop("FullDate", axis=1)
```

```python
df["time"] = (df["time"]).astype(str)
df["time"] = df['time'].apply(lambda x: x[:2])
df["time"] = (df["time"]).astype(int)
```

```python
df.info()
```

```python
df.drop(columns=['DateTimeCol'], inplace=True)

# Display the resulting DataFrame after dropping the column
print(df)
```

```python
#Resample to daily data points
df_daily = df.resample('D').sum()

plt.figure(figsize=(10,6))
plt.plot(df_daily, '--', marker='*')
plt.grid()
plt.xlabel('date')
plt.ylabel('GasPrice')
plt.title('Upsampled Data by Day')
```
